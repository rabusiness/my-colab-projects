```python
#Numpy fitness tracker tutorial

#importing libraries
import numpy as np
import matplotlib.pyplot as plt

#Section 1: NumPy Arrays

#creating arrays from lists
daily_steps = np.array([8500,7400,10000,12000,9000,0,17000])
daily_calories = np.array([1200,950,1800,2300,1500,0,2400])
workout_duration = np.array([45,30,60,75,50,0,90])

print("Daily steps take each day of week:", daily_steps)

#Array properties
print("Shape of daily_steps array:", daily_steps.shape)
print("Data type of daily_steps array:", daily_steps.dtype)
print("Number of dimensions of daily_steps array:", daily_steps.ndim)
print("size of daily_steps array:", daily_steps.size)

#Create array with specific values
heart_rate_zones = np.zeros(5)
print("Heart rate zones(initiated to zero):", heart_rate_zones)

weekly_goal_met = np.ones(7)
print("Weekly goals met(all day):", weekly_goal_met)

#create array with range
days_of_month = np.arange(1,32)
print("days of month:", days_of_month)

#create array with specific spacing
#np.linspace(start,end,how many evenly spaced values?): creates an array of numbers starting from
workout_weeks = np.linspace(1,12,12)
print("workout weeks:", workout_weeks)
```

```python
#Section 3

#Element-wise operations
print("Steos + 1000 bonus:", daily_steps + 1000)
print("Steps * 1.2:", daily_steps * 1.2)
print("Calories/100", daily_calories/100)

total_activity_score = daily_steps/100 + workout_duration
print("Total activity score:", total_activity_score)
```

```python
#Section 6: multi-dimensional arrays(weekly data): 4 weeks x 7 days

monthly_steps = np.array([
    [8500, 9200,7500,10000,9000,6500,12000],
    [7200,8900,8100,9500,8880,7200,10500],
    [9100,9800,8400,11200,10100,8900,12000],
    [8800,9500,7900,10800,9200,7800,11500]
])

print("Monthly steps data shape:", monthly_steps.shape)
print("Average steps per week:")
weekly_averages = np.mean(monthly_steps, axis=0)
# Calculate the overall average of the weekly averages and format it
overall_average_steps = np.mean(weekly_averages)
print(f"Overall average daily steps: {overall_average_steps:.2f}")
print(f"Average steps per week: {weekly_averages}")
```

```python
def calculate_bmi(weight_kg, height_m):
  bmi = weight_kg / (height_m ** 2)
  return bmi
def calories_burned(daily_steps, weight_kg):
  return daily_steps * 0.04 * weight_kg

def fitness_level(weekly_avg_steps):
  choices = ['Sedentary', 'Low active', 'Moderaely Active', 'Active']
  conditions = [
      weekly_avg_steps < 5000,
      (weekly_avg_steps >= 5000) & (weekly_avg_steps < 10000),
      (weekly_avg_steps >= 10000) & (weekly_avg_steps < 20000),
      weekly_avg_steps >= 20000
  ]
  return np.select(conditions, choices, default = 'Unknown')

weights = np.array([70,65,80,75])
heights = np.array([1.75, 1.68, 1.82, 1.70])
bmis = calculate_bmi(weights,heights)
print("BMI values:", bmis.round(1))

estimated_calories = calories_burned(daily_steps, 70)
print("Estimated calories burned:", estimated_calories)

weekly_steps_average = np.mean(monthly_steps, axis=1)
print("Weekly steps average:", weekly_steps_average)
fitness_levels = fitness_level(weekly_averages)
print("Fitness levels:", fitness_levels)
```

```python
#section: visualisation

import matplotlib.pyplot as plt

plt.figure(figsize=(15,12))

#Plot 1: daily steps line chart
plt.subplot(2,3,1)
days_of_week = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'] # Define days of the week
days = range(len(daily_steps)) # Use range starting from 0 for indexing
plt.plot(days,daily_steps,marker='o',linestyle='-',color='b')
plt.title('Daily Steps')
plt.xlabel('Day')
plt.ylabel('Steps')
plt.xticks(days, days_of_week, rotation=45) # Set x-ticks and labels
plt.grid(True,alpha = .3)
plt.tight_layout()

#Plot 2: Daily Calories bar chart
plt.subplot(2,3,2)
colors = ['lightgreen' if cal >= 2200 else 'lightcoral' for cal in daily_calories]
bars = plt.bar(days, daily_calories, color=colors, alpha=0.7)
plt.title('Daily Calories')
plt.xlabel('Day')
plt.ylabel('Calories')
plt.xticks(days, days_of_week, rotation=45) # Set x-ticks and labels
plt.tight_layout()

# Plot 3: Daily workout duration bar chart
plt.subplot(2, 3, 3)
duration_colors = ['yellow' if duration > 40 else 'blue' for duration in workout_duration]
plt.bar(days, workout_duration, color=duration_colors, alpha=0.6)
plt.title('Daily Workout Duration')
plt.xlabel('Day')
plt.ylabel('Duration (minutes)')
plt.xticks(days, days_of_week, rotation=45)
plt.tight_layout()
```
